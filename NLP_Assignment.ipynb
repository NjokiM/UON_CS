Write code to search the Brown Corpus 
(https://www.kaggle.com/datasets/nltkdata/brown-corpus) for particular  words and phrases according to tags, to answer the following questions: 





a) Produce an alphabetically sorted list of the distinct words tagged as  MD

b) Identify words that can be plural nouns or third person singular  verbs (e.g., deals, flies).  
c) Identify three-word prepositional phrases of the form IN + DET +  NN (e.g., in the lab).  
d)  What is the ratio of masculine to feminine pronouns? 





```python
import nltk
from nltk import word_tokenize
from nltk.corpus import brown
nltk.download('brown')
from nltk.probability import FreqDist
nltk.download('averaged_perceptron_tagger')
import pandas as pd
nltk.download('punkt')



```

```python
text = "They wind back the clock,  while we chase after the wind"
token_text = word_tokenize(text)
print(nltk.pos_tag(token_text), end = '')
```

```python
# Load the Brown Corpus
corpus = brown.tagged_words()

# Filter the corpus to only include words tagged as "MD"
md_words = [word.lower() for (word, tag) in corpus if tag == 'MD']

# Get a sorted list of distinct words
distinct_words = sorted(set(md_words))

# Print the list of words
print(distinct_words)
```

```python
# Define the tags for plural nouns: 'NNS', 'NNPS' and third person singular verbs:'VBZ'
pn_and_tp = ['NNS','NNPS','VBZ']

# Loop through the Brown Corpus and identify words that match the search words and tags
pn_and_tp_words = [word.lower() for (word, tag) in corpus if tag in pn_and_tp and word.isalnum]
distinct_pn_and_tp_words = set(pn_and_tp_words)
print(FreqDist(distinct_pn_and_tp_words))
```

```python

def find_plural_nouns_and_3rd_person_verbs():
    plural_nouns = []
    third_person_verbs = []

    # Get all the words in the Brown Corpus
    words = brown.words()

    # Tag each word with its part-of-speech using the Brown Corpus tagger
    tagged_words = nltk.pos_tag(words)

    # Iterate through the tagged words and identify plural nouns and third person singular verbs
    for word, tag in tagged_words:
        if tag == 'NNS':  # NNS tag represents plural nouns
            plural_nouns.append(word)
        elif tag == 'VBZ':  # VBZ tag represents third person singular verbs
            third_person_verbs.append(word)

    return plural_nouns, third_person_verbs

```

```python
# Call the function to find plural nouns and third person singular verbs
plural_nouns, third_person_verbs = find_plural_nouns_and_3rd_person_verbs()

# Print the results
print("Plural Nouns:")
print(plural_nouns[:10])  # Print the first 10 plural nouns
print()

print("Third Person Singular Verbs:")
print(third_person_verbs[:10]) 
```

```python
def find_prepositional_phrases():
    prepositional_phrases = []
    tagged_sentences = brown.tagged_sents(categories='news')  # Use 'news' category of Brown Corpus

    for sentence in tagged_sentences:
        for i in range(len(sentence) - 2):
            if sentence[i][1].startswith('IN') and sentence[i + 1][1].startswith('AT') and sentence[i + 2][1].startswith('NN'):
                phrase = ' '.join([sentence[i][0], sentence[i + 1][0], sentence[i + 2][0]])
                prepositional_phrases.append(phrase)

    return prepositional_phrases

# Example usage
phrases = find_prepositional_phrases()
for phrase in phrases:
    print(phrase)
```

```python
# Load the Brown Corpus
brown_corpus = nltk.corpus.brown

# Get all of the masculine pronouns
masculine_pronouns = [
    "he",
    "him",
    "his",
    "himself",
]

# Get all of the feminine pronouns
feminine_pronouns = [
    "she",
    "her",
    "hers",
    "herself",
]

# Count the number of masculine and feminine pronouns
masculine_count = 0
feminine_count = 0

for sentence in brown_corpus.tagged_sents():
    for word, tag in sentence:
        if word in masculine_pronouns:
            masculine_count += 1
        elif word in feminine_pronouns:
            feminine_count += 1

# Calculate the ratio of masculine to feminine pronouns
ratio = masculine_count / feminine_count

print(ratio)
```

```python

corpus = nltk.corpus.brown

# Define lists of masculine and feminine pronouns
masculine_pronouns = ['he', 'him', 'his', 'himself']
feminine_pronouns = ['she', 'her', 'hers', 'herself']

# Initialize counts for masculine and feminine pronouns
masculine_count = 0
feminine_count = 0

# Iterate over the tagged words in the corpus
for word, tag in corpus.tagged_words():
  if word.lower() in masculine_pronouns:
      masculine_count += 1
  elif word.lower() in feminine_pronouns:
      feminine_count += 1

# Calculate the ratio of masculine to feminine pronouns
ratio = masculine_count / feminine_count

# Print the result
print(f"The ratio of masculine to feminine pronouns in the Brown Corpus is {ratio:.2f}.")
```
